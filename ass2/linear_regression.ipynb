```python
import pandas as pd
from scipy.io import arff
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import explained_variance_score
from matplotlib import *
from sklearn.metrics import accuracy_score
import operator
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
filename = "autoMpg.arff"
data = arff.loadarff(filename)
column_names = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model','origin', 'class']

#df = pd.read_csv(filename, sep=',', names=column_names)
df = pd.DataFrame(data[0], columns=column_names)
for i in df.columns:
    df[i] = df[i].fillna(df[i].mean())
#print(df)
Cylinders = [8, 4, 6, 3, 5]


def label_race(row, col_name, value):
    #print(int(row[col_name]))
    if float(row[col_name]) == value:
        #print('aaaa')
        return 1
    else:
        return 0


for value in Cylinders:
    df['cylinder_' + str(value)] = df.apply(lambda row: label_race(row, 'cylinders', value), axis=1)

Model = [70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 81, 82]

for value in Model:
    df['model_' + str(value)] = df.apply(lambda row: label_race(row, 'model', value), axis=1)

Origin = [1, 2, 3]

for value in Origin:
    df['origin_' + str(value)] = df.apply(lambda row: label_race(row, 'origin', value), axis=1)

df=df.drop(columns=['cylinders', 'model', 'origin'],axis=1)
cols = df.columns.tolist()
cols=cols[:3]+cols[5:]
cols.append('class')
df = df[cols]
y = df['class']
df=df.drop(columns=['class'],axis=1)
x=df.loc[:, 'displacement':'origin_3']
#sc = StandardScaler()
#sc.fit(x)


weights = [0.001]*(x.shape[1]+1)
#print(weights)
#print(x.shape[1])
def predict(row, weights):
	activation = weights[0]
	for i in range(df.shape[1]-1):
		activation += weights[i+1] * row[i]
	return activation
def linear_regression(x, y, weights, epochs=453, learning_rate=0.0000001043):
    N = float(len(y))
    
    for i in range(epochs):
        cost=0
        prediction=[]
        diff=[]
        best_iter=0
        score={}
        pprediction=[]
        pweights=[0.001]*(x.shape[1]+1)
        for j in range(len(y)):
            prediction.append(predict(x.iloc[j],weights))
            y_current = prediction
            diff.append(y[j]-y_current[j])
            cost +=(prediction[j]-y[j])**2 
        sum_diff=sum(diff)
        weights[0]=weights[0]-learning_rate*(-(2/N) * sum_diff)
        for k in range(x.shape[1]-1):
            weights[k+1]=weights[k+1]-learning_rate*(-(2/N) * np.dot(x.iloc[:,k] , (y - y_current)))
        cost=cost/N
        var=explained_variance_score(y,prediction)
        score[i]=var
        print(i,score[i])
    return i,score[i],weights,prediction
best_iter,score,weights,prediction=linear_regression(x,y,weights)
```
