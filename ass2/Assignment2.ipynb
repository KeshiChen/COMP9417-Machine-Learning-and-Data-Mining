```python
import pandas as pd
filename = "auto-mpg.txt"
column_names = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'Model year','Origin', 'mpg']
df = pd.read_csv(filename, sep=',', names=column_names)

print (df.shape)
df.head(5)
```

```python

horsepower_missing_ind = df[df.horsepower=='?'].index
df.loc[horsepower_missing_ind]
Cylinders=[8,4,6,3,5]

def label_race (row,col_name,value):
    if row[col_name] == value :
        return 1
    else:
        return 0
for value in Cylinders:
    df['Cylinder_'+str(value)]=df.apply (lambda row: label_race (row,'cylinders',value),axis=1)

Model=[70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 81, 82]

for value in Model:
    df['Model_'+str(value)]=df.apply (lambda row: label_race (row,'Model year',value),axis=1)
    
Origin=[1,2,3]

for value in Origin:
    df['Origin_'+str(value)]=df.apply (lambda row: label_race (row,'Origin',value),axis=1)
df
```

```python

df.loc[horsepower_missing_ind, 'horsepower'] = float('nan')
df.horsepower = df.horsepower.apply(pd.to_numeric)
df.loc[horsepower_missing_ind, 'horsepower'] = int( df.horsepower.mean() )
pd.set_option('precision', 2)
#drop columns which are multi-valued
df=df.drop(['cylinders','Model year','Origin'],axis=1)
#rearrange column position
cols = df.columns.tolist()
cols=cols[:3]+cols[5:]
cols.append('mpg')
df = df[cols]
#df
df
```
